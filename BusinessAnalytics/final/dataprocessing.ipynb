## Merge arm-eng columns
* Add one column which indicates whether the information was filled out in Armenian or English
* Add one more column which indicates whether the candidate was accepted
* Translation / Transliteration will be handled later

```python
df[['Էլեկտրոնային հասցե (e-mail) ', 'E-mail address']]
```

```python
# Put the processed data in a new dataframe
d = pd.DataFrame()
```

```python
arm_columns = df.columns[: len(df.columns) // 2]
eng_columns = df.columns[len(df.columns) // 2: ]
```

```python
def merge_columns(row):
    res = row.copy()
    res['language'] = 'eng' if row['First Name'] else 'arm'
    for arm, eng in zip(arm_columns, eng_columns):
        del res[arm]
        res[eng] = row[arm] or row[eng]
    return res
```

```python
d = df.apply(merge_columns, axis=1)
d.head()
```

```python
def process_email(email):
    email = email.lower()
    email.replace(',', ' ')
    email.replace(';', ' ')
    email.replace('\n', ' ')
    return email.split()[0]
```

```python
d['email'] = d.apply(lambda x: process_email(x['E-mail address']), axis=1)
del d['E-mail address']
```

## Remove the variables which don't have anything to do with acceptance

```python
del d['I hereby give my consent to ']
del d['Mobile phone number']
del d['First Name']
del d['Last Name']
```

## Add the column which indicates wheter the candidate was accepted

```python
acc = pd.read_csv('accepted.csv')
acc = acc.where((pd.notnull(acc)), None)
acc.head()

acc['email'] = acc.apply(lambda x: process_email(x['Email']), axis=1)
del acc['Email']
```

```python
d['accepted'] = False
```

```python
for i, person in tqdm(acc.iterrows()):
    email = person['email']
    d.ix[d['email'] == email, ['accepted']] = True
```

```python
for i, row in d.iterrows():
    if row['language'] == 'eng':
        continue

    for col in categorical:
        if row[col] is None:
            continue
        translate = []
        for armenian in hy2en.keys():
            if armenian in row[col]:
                translate.append(hy2en[armenian])
        row[col] = ','.join(translate) if translate else row[col]
    
    d.at[i] = row
```

## As we can see, we are left with only open-ended questions
* The answers to those can be either left as is or translated to english
* As we may use fastText embeddings as features, we are going to translate the answers
* The vectors from fastText for different languages are not comparable so there is no way of using armenian and english vectors in the same model
* We could've used mechanisms like MUSE to transform the Armenian vectors to English space, but that's much more work compared to translation

## The translation can be done through the Google API
* But as most of the services are not free after some limit
* We will use google sheets to do the translation for us
* Use `=ArrayFormula(GOOGLETRANSLATE(D2:D, "auto", "en"))` as a formula for a cell to translate the column `D`
* Apply this step for all the columns in the data and obtain english translations
